## CÂU 1 
### Cho dữ liệu BMI.xlsx. Bộ dữ liệu này ghi lại BMI của một số quốc gia trong một số năm
Đọc dữ liệu 
Xem xét vấn đề về dữ liệu cần khắc phục 
Chuẩn lại dữ liệu để khắc phục vấn dề trên 

Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo country, year. 
Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

```python
import numpy as numpy
import pandas as pd
```

```python
df = pd.read_excel('BMI.xlsx')
df.head()
```

```python
# Vấn đề cần khắc phục: tên cột chứa giá trị thay vì chứa biến
# Melt df new dataframe: df.melt()
df_melt = pd.melt(df,id_vars='Country',var_name='Year',value_name='BMI')
df_melt
```

```python
df_melt['Year'] = df_melt['Year'].apply(lambda x: x[1:])
df_melt
```

```python
# Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo country, year. 
# Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê
df_melt_pivot = df_melt.pivot(index='Country',columns='Year',values='BMI')
df_melt_pivot
```

# Câu 2: 
* Cho dữ liệu student.xlsx. Bộ dữ liệu này ghi lại điểm các môn của sinh viên 
* Đọc dữ liệu 
* Xem xét vấn đề về dữ liệu cần khắc phục 
* Chuẩn lại dữ liệu để khắc phục vấn dề trên 

* Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo từng student và các subject. 
* Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

```python
df = pd.read_excel('student.xlsx')
df.head()
```

```python
# Melt chuyển đổi cột giá trị thay vì cột biến
df_melt = pd.melt(df,id_vars='Student',var_name='Môn học',value_name='Điểm')
df_melt 
```

```python
df_melt_pivot = df_melt.pivot(index='Student',columns='Môn học',values='Điểm')
df_melt_pivot
```
