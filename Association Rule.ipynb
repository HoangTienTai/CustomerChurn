## Step1: Data Raw

* Transaction: Giao dịch

* Trong giao dịch thì sẽ có các item

```python
ls_trans =[
["Apple", "Beer", "Rice", "Chicken"],
["Apple", "Beer", "Rice"],
["Apple", "Beer"],
["Apple", "Pear"],
["Milk", "Beer", "Rice", "Chicken"],
["Milk", "Beer", "Rice"],
["Milk", "Beer"],
["Milk", "Pear"]
]
print(ls_trans)
# Transaction: Giao
```

## Step 2: Xử lý Data Raw -> bảng Transaction

* Sử dụng thư viện TransactionEncoder

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
model_transaction = TransactionEncoder()
ls_trans = model_transaction.fit_transform(ls_trans)
df_trans = pd.DataFrame(ls_trans,columns = model_transaction.columns_)
df_trans.replace(['True','False'],[1,0], inplace = True)
df_trans
```

## Step3: Using Apriori to count Support

```python
from mlxtend.frequent_patterns import apriori
df_support = apriori(df_trans,min_support = 0.01,use_colnames=True)
df_support = df_support.sort_values("support",ascending=False)
df_support
```

## Step 4: Using Association Rule to calculate Lift, Support, Confidence

=> Association rule chỉ hỗ trợ mn đưa ra quyết định thôi

```python
from mlxtend.frequent_patterns import association_rules
df_association = association_rules(df_support,metric='lift',min_threshold=1)
# antecedents: A (xuất hiện trước)
# consequents: B (Xuất hiện sau) A-> B
df_association.sort_values(["support","confidence","lift"],ascending=False)
```
