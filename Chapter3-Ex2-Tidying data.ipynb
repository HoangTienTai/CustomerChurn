# Câu 1: Pew Research Center 
* Cho dữ liệu: pew-raw.csv. Bộ dữ liệu này khám phá mối quan hệ giữa thu nhập và tôn giáo (income & religion) 
* Đọc dữ liệu 
* Xem xét vấn đề về dữ liệu cần khắc phục 
* Chuẩn lại dữ liệu để khắc phục vấn đề trên

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pew-raw.csv')
```

```python
df.head()
```

```python
# Vấn đề về dữ liệu cần khắc phục: cột giá trị thay vì cột tên biến
df_melt = pd.melt(df,id_vars='religion',var_name='Income',value_name='Price')
df_melt.head()
```

```python
df_melt['Income'] = df_melt['Income'].str.replace('$','')
df_melt.head()
```

```python
# Bảng thống kê dữ liệu lại bằng pivot
df_melt_pivot = df_melt.pivot(index='religion',columns='Income',values='Price')
df_melt_pivot.head()
```

# Câu 2: Billboard Top 100 
* Cho dữ liệu billboard.csv. Bộ dữ liệu này đại diện cho thứ hạng hàng tuần của các bài hát kể từ thời điểm chúng lọt vào Top 100 của Billboard cho * đến 75 tuần tiếp theo. 
* Đọc dữ liệu 
* Xem xét vấn đề về dữ liệu cần khắc phục 
* Chuẩn lại dữ liệu để khắc phục vấn dề trên 
* Sau khi chuẩn lại dữ liệu, có điều gì phát sinh cần khắc phục tiếp theo không? Nếu có thì hãy chuẩn lại dữ liệu mới có ở câu trên

```python
df = pd.read_csv('billboard.csv',encoding='mac_latin2')
df.head()
```

```python
# Melt
df_melt = pd.melt(df,id_vars=['year','artist.inverted','track','time','genre','date.entered','date.peaked'],var_name='Week',value_name='Rank')
df_melt.head()
```

```python
df_melt.tail()
```

```python
df_melt.info()
```

# Cần chuẩn hóa dữ liệu
* Thay giá trị của Week từ chuỗi thành số
* Xóa bỏ các dòng mà cột Rank có Nan
* Chuyển kiểu dữ liệu của Rank thành kiểu int

```python
df_melt['Week'] = df_melt['Week'].str.extract('(\d+)',expand=False)
df_melt
```

```python
df_melt = df_melt.dropna()
df_melt.tail()
```

```python
df_melt['Rank'] = df_melt['Rank'].astype(int)
```

```python
df_melt.info()
```

```python
df_melt.head()
```

## Câu 3 tb.raw.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df3 = pd.read_csv('tb-raw.csv')
```

```python
df3.head()
```

```python
df3.info()
```

#### Vấn đề cần khắc phục
* Tên cột chứa giá trị thay vì chứa biến
* Trong ce11 chứa giá trị hỗn hợp gồm value, 0, NaN

```python
df3_new = pd.melt(df3,id_vars=['country','year',],var_name='sex and age',value_name='cases')
df3_new.head()
```

```python
df3_new = df3_new.dropna()
df3_new.tail()
```

### Vấn đề phát sinh
* trong cùng một cột chứa 2 thông tin độ tuổi và giới tính

```python
df_temp = df3_new['sex and age'].str.extract('(\D)(\d+)(\d{2})')
```

```python
df_temp.head()
```

```python
df_temp.columns = ['Sex','Age_lower','Age_upper']
df_temp['Age'] = df_temp['Age_lower'] + "-" + df_temp['Age_upper'] 
df_temp.head()
```

```python
df3_new = pd.concat([df3_new,df_temp],axis=1)
df3_new.head()
```

```python
df3_new = df3_new[['country','year','Sex','Age','cases']]
df3_new.head()
```
