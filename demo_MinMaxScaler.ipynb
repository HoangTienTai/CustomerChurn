```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Data_for_Values.csv')
data.head()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.distplot(data['Age'])
plt.subplot(1,2,2)
sns.distplot(data['Salary'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(data['Age'])
plt.subplot(1,2,2)
sns.boxplot(data['Salary'])
plt.show()
```

```python
# Sử dụng MinMax Scaler vì không có outlier
from sklearn.preprocessing import MinMaxScaler
```

```python
# Khởi tạo
scale = MinMaxScaler()
df_new = data[['Age','Salary']]
df_scale = scale.fit_transform(df_new)
df_scale = pd.DataFrame(df_scale,columns=df_new.columns)
df_scale
```

```python
f, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,6))
sns.kdeplot(df_new['Age'], ax=ax1,label = 'Age')
sns.kdeplot(df_new['Salary'],ax=ax1,label ='Salary')
ax1.set_title('Before Scaling')
sns.kdeplot(df_scale['Age'],ax=ax2,label = 'Age')
sns.kdeplot(df_scale['Salary'],ax=ax2,label = 'Salary')
ax2.set_title('After Scaling')
plt.legend()
plt.show()
```

```python
print('Age before scaling:', df_new['Age'].min(), df_new['Age'].max())
print('Age after scaling:', df_scale['Age'].min(), df_scale['Age'].max())
```

```python
print('Salary before scaling:', df_new['Salary'].min(), df_new['Salary'].max())
print('Salary after scaling:', df_scale['Salary'].min(), df_scale['Salary'].max())
```
