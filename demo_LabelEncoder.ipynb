```python
import pandas as pd
```

```python
df = pd.read_csv('ds_nhanvien.csv')
df.head()
```

```python
# Mã hóa label cột vị trí, chuyển chữ thành số
from sklearn.preprocessing import LabelEncoder
# Khởi tạo
encoder = LabelEncoder()
df['Vitri_Encoder'] = encoder.fit_transform(df['Vitri'])
df.head()
```

```python
# label encoder sắp xếp các giá trị theo bảng chữ cái rồi mới đánh số.
```
