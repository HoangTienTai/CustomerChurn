```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mammals.csv')
data.head()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data['BodyWt'],data['BrainWt'])
plt.xlabel('Body Weight')
plt.ylabel('Brain Weight')
plt.show()
```

```python
data['BodyWt_log'] = np.log(data['BodyWt'])
data['BrainWt_log'] = np.log(data['BrainWt'])
```

```python
data.head()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data['BodyWt_log'],data['BrainWt_log'])
plt.xlabel('Body Weight log')
plt.ylabel('Brain Weight log')
plt.show()
```
