```python
# 1 Tạo DataFrame df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. Các danh sách
# này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv
```

```python
# Import các thư viện để tiến hành phân tích
import numpy as np
import pandas as pd
```

```python

# 1 Đọc các tập tin csv chứa danh sách sinh viên từ các khoa
df_CN = pd.read_csv('sinh_vien_CN.csv')
df_TO = pd.read_csv('sinh_vien_TO.csv')
df_VL = pd.read_csv('sinh_vien_VL.csv')

# Tạo DataFrame df chứa danh sách sinh viên từ các khoa
df = pd.concat([df_CN, df_TO, df_VL])

# In ra DataFrame df
print(df)
```

```python
# Kiểu dữ liệu của DataFrame df
df.info()
```

```python
# 2 Cho biết số cột, số dòng, các dữ liệu head và tail của DataFrame df.
```

```python

# Số cột của DataFrame df
so_cot = df.shape[1]

# Số dòng của DataFrame df
so_dong = df.shape[0]

# Dữ liệu head của DataFrame df (5 dòng đầu tiên)
head_data = df.head()

# Dữ liệu tail của DataFrame df (5 dòng cuối cùng)
tail_data = df.tail()

# In kết quả
print("Số cột:", so_cot)
print("Số dòng:", so_dong)
print("Dữ liệu head:")
print(head_data)
print("Dữ liệu tail:")
print(tail_data)
```

```python
# 3 Tạo lại index cho DataFrame df với giá trị bắt đầu từ 0
```

```python

# Tạo lại index với giá trị bắt đầu từ 0
df = df.reset_index(drop=True)

# In ra DataFrame df đã được tạo lại index
print(df)
```

```python
# 4 Tạo DataFrame df_ket_qua từ tập tin ket_qua.csv
```

```python

# Đọc tập tin csv và tạo DataFrame df_ket_qua
df_ket_qua = pd.read_csv('ket_qua.csv')

# In ra DataFrame df_ket_qua
print(df_ket_qua)
```

```python
# 5 Tạo DataFrame df_dtb bằng cách nhóm df_ket_qua theo masv và tính điểm trung bình cho mỗi sinh viên.
```

```python
# Nhóm df_ket_qua theo masv và tính điểm trung bình
df_dtb = df_ket_qua.groupby('masv').mean()

# In ra DataFrame df_dtb
print(df_dtb)
```

```python
# 6 Trộn df và df_dtb theo masv và có kết quả của df như sau:
# Trộn df và df_dtb theo masv
df_merged = pd.merge(df, df_dtb, on='masv')

# In ra kết quả của df_merged
print(df_merged)
```

```python
# 7 In ra các sinh viên của khoa CN và có diem>8
# Lọc các sinh viên của khoa CN và có điểm > 8
df_filtered = df_merged[(df_merged['makh'] == 'CN') & (df_merged['diem'] > 8)]

# In ra các sinh viên thỏa điều kiện
print(df_filtered)
```

```python
# 8 In ra các sinh viên của khoa CN và có diem>8. Sắp xếp tăng theo ten và giảm theo diem
# Sắp xếp tăng dần theo tên và giảm dần theo điểm
df_sorted = df_filtered.sort_values(by=['ten', 'diem'], ascending=[True, False])

# In ra các sinh viên thỏa điều kiện và đã được sắp xếp
print(df_sorted)
```

```python
# 9 In ra các nam sinh viên không có học bổng.
# Lọc các sinh viên nam không có học bổng
# True là nam, False là nữ theo dữ liệu trong file.

df_filtered = df_merged[(df_merged['gioitinh'] == True) & (df_merged['hocbong'] == 0)]

# In ra các sinh viên thỏa điều kiện
print(df_filtered)
```

```python
# 10 In ra tổng học bổng của mỗi khoa.
# Tính tổng học bổng của mỗi khoa
hoc_bong_theo_khoa = df_merged.groupby('makh')['hocbong'].sum().reset_index()
# In ra tổng học bổng của mỗi khoa
print(hoc_bong_theo_khoa)
```

```python
# 11 Thống kê học bổng của mỗi khoa.
thong_ke_hoc_bong = df_merged.groupby('makh')['hocbong'].agg(['sum','min','max','mean']).reset_index()
thong_ke_hoc_bong
```

```python
# 12 Thống kê điểm của mỗi khoa
thong_ke_diem = df_merged.groupby('makh')['diem'].agg(['count','min','max','mean']).reset_index()
thong_ke_diem
```

```python
# 13 Cho biết các sinh viên có điểm lớn nhất, điểm nhỏ nhất.
# Lấy ra số điểm lớn nhất
diem_lon_nhat = df_merged['diem'].max()

# Lấy ra số điểm nhỏ nhất
diem_nho_nhat = df_merged['diem'].min()

# Hiển thị sinh viên có số điểm lớn nhất
sinh_vien_diem_max = df_merged[df_merged['diem'] == diem_lon_nhat]
print('Sinh viên có số điểm lớn nhất\n')
print(sinh_vien_diem_max)

# Hiển thị sinh viên có số điểm nhỏ nhất
sinh_vien_diem_min = df_merged[df_merged['diem'] == diem_nho_nhat]
print('Sinh viên có số điểm nhỏ nhất\n')
print(sinh_vien_diem_min)
```

```python
# 14 Trong DataFrame df tạo thêm cột ketqua với kết quả là giỏi nếu điểm>=8 và kết quả là khá nếu điểm<8.
# Tạo hàm để áp dụng cho mỗi giá trị trong cột "diem"
def ketqua(diem):
    if diem>=8:
        return 'giỏi'
    else:
        return 'khá'
df_merged['ketqua'] = df_merged['diem'].apply(ketqua)
# df_merged['ketqua'] = df_merged['diem'].apply(lambda x: ketqua(x))
df_merged
```

```python
# 15 Thống kê theo mẫu:
df_theo_mau = df_merged[['makh','ketqua','masv']]
df_theo_mau
```
